```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
transaction_data = pd.read_csv('transaction_data.csv')
```

```python
transaction_data.dtypes
```

```python
missing_values_per_column = transaction_data.isnull().sum()

# Найдем общее количество пропущенных значений во всем датасете
total_missing_values = transaction_data.isnull().sum().sum()

print(missing_values_per_column)
print("Общее количество пропущенных значений в датасете:", total_missing_values)
```

```python
transaction_data.describe
```

```python
# Найти уникальные значения и количество наблюдений для колонки 'transaction'
value_counts = transaction_data['transaction'].value_counts()

value_counts
```

```python
# Визуализировать результаты
plt.figure(figsize=(10, 6))
sns.barplot(x=value_counts.index, y=value_counts.values)
plt.xlabel('Transaction Types')
plt.ylabel('Number of Observations')
plt.title('Number of Observations per Transaction Type')
plt.xticks(rotation=45)  
plt.tight_layout()
```

```python
# Подсчитываем количество транзакций с ошибкой
error_transactions = transaction_data[transaction_data['transaction'] == 'Error'].shape[0]

error_transactions
```

```python
# Фильтруем успешные транзакции
successful_transactions = transaction_data[transaction_data['transaction'] == 'successfull']

# Группируем по 'user_id' и подсчитываем количество успешных транзакций
transactions_per_user = successful_transactions.groupby('name').size()

transactions_per_user

```

```python
plt.figure(figsize=(10, 6))
transactions_per_user.hist(bins=50)
plt.title('Распределение числа успешных транзакций на пользователя')
plt.xlabel('Количество успешных транзакций')
plt.ylabel('Количество пользователей')
plt.grid(True)
plt.show()
```
